```

stride = 2
HH =4
H=4
W=4
WW =4
pad =1
N=2
F=3

import numpy as np  
w_shape = (3, 3, 4, 4)
np.set_printoptions(threshold=np.inf) 
w = np.linspace(-0.2, 0.3, num=np.prod(w_shape)).reshape(w_shape)
b = np.linspace(-0.1, 0.2, num=3)


H_out = int(1 + (H + 2 * pad - HH) / stride)
print("H_out :",H_out)
W_out = int(1 + (W + 2 * pad - WW) / stride)
print("W_out :",W_out)
out = np.zeros((N, F, H_out, W_out))
print("x : ",x)
#zero padding      前面的數字 代表在該維度 要不要加0
#                               row      col
x_pad = np.pad(x,((0,), (0,), (pad,), (pad,)) ,mode='constant', constant_values=0)
#                                       前面1代表col 的前面   後面1 代表col 後面
print("x_pad:",x_pad)
for i in range(H_out):
       for j in range(W_out):
          #跳stride
          # 取出filter的大小 
          #                                    : 假如是4 則只會從0~3 
          x_padded_mask = x_pad[:, :, i*stride:i*stride+HH, j*stride:j*stride+WW] #(:, :, HH, WW)
          print(x_padded_mask)
          for k in range(F): # F =3   k 只會做到2
          #卷积
            print("k :",k)
            print("out.shape :",out.shape)  # (2, 3, 2, 2))
            print("x_padded_mask.shape :",x_padded_mask.shape) # (2, 3, 4, 4)
            #                         每個filter 對 每一張照片 RGB 取出同樣大小的矩陣  把三個filter 取出來的值 全部sum起來  成一個數值 接者依序放入 out矩陣內 
            out[:, k , i, j] = np.sum(x_padded_mask * w[k, :, :, :], axis=(1,2,3))   #    如果第一個維度有兩張照片  假設我沒寫0  代表第一個row 我不做相加 
print("out ",out)         
out += b[None, :, None, None]# None 相当于 numpy.newaxis  這樣才能確實與out 各個相加
print("b[None, :, None, None] :",b[None, :, None, None])

```

```
global h
h = 0.00001
eps = 0.0001

# f(x) to solve
def f(x):
    a_0 = 1.00000
    a_1 = 2.00000
    a_2 = 3.00000
    a_3 = 4.00000
    a_4 = 5.00000
    return a_0 + a_1 * x + a_2 * x*x + a_3 * x*x*x + a_4 * x*x*x*x

def fp(x):
    global h
    return (f(x + h) - f(x)) / h

def fpp(x):
    global h
    return (fp(x + h) - fp(x)) / h

# main
x = 2 # initial value

while True:
    fx = f(x)
    fpx = fp(x)
    xnew = x - (2.0 * fx * fpx) / (2.0 * fpx * fpx - fx * fpp(x))
    #print(xnew)
    if abs(xnew - x) <= eps: 
        print(xnew,"\n\ndone!")
        break
    x = xnew
```
